#Feature info

```python
"""
Patient_ID – unique identifier for each patient.

Visit_ID – unique identifier for each clinical visit (patients can have multiple visits).

Date_of_birth – patient’s date of birth.

Symptom_onset_date – date when symptoms first appeared.

Diagnosis_date – date of confirmed diagnosis.

Sex – patient’s gender.

Age_at_onset – age when symptoms started.

Clinical_onset_type – type of symptom onset (e.g., limb, bulbar, respiratory).

Spirometry – lung function measure (often FVC%).

FVC – Forced Vital Capacity (key respiratory measure).

Ventilation – whether patient uses ventilatory support (yes/no).

Tracheostomy – whether patient has undergone tracheostomy.

PEG – whether patient has a percutaneous endoscopic gastrostomy (feeding tube).

Family_history – family history of the disease (yes/no).

Therapy – treatments patient is receiving.

Diagnostic_delay – time between symptom onset and diagnosis (months).

MRC_Upper_Limb – Medical Research Council score for upper limb strength.

MRC_Lower_Limb – MRC score for lower limb strength.

MRC_Bulbar – MRC score for bulbar function (speech/swallowing).

PUMNS_Upper_Limb – Penn Upper Motor Neuron Score for upper limb.

PUMNS_Lower_Limb – Penn UMN Score for lower limb.

PUMNS_Bulbar – Penn UMN Score for bulbar region.

CNS_Linear_Score – Composite Neurological Score (linear scale).

KINGS_Total – King’s clinical staging score (disease progression stage).

Disease_duration – time since symptom onset (months or years).

Progression_rate – rate of disease worsening (e.g., ALSFRS decline per month).

ALSFRS_R – Revised ALS Functional Rating Scale (0–48; higher = better function).

ALSAQ_5 – ALS Assessment Questionnaire, 5-item summary (quality of life).
"""
```

##Spirometry

```python
print(df['Spirometry'].unique())
```

```python
print(df['Spirometry'].value_counts())
```

```python
# Filter the DataFrame to find rows where 'Spirometry' is null
spirometry_null_patients = df[df['Spirometry'].isnull()]

# Display the Patient_IDs for those rows
display(spirometry_null_patients['Patient_ID'])
```

```python
# Display the Patient_ID and Spirometry columns for rows with null Spirometry values
display(spirometry_null_patients[['Patient_ID', 'Spirometry']])
```

```python
patient_ids_with_null_spirometry = df[df['Spirometry'].isnull()]['Patient_ID'].unique()
display(patient_ids_with_null_spirometry)
```

```python
rows_with_null_spirometry = df[df['Patient_ID'].isin(patient_ids_with_null_spirometry)][['Patient_ID', 'Spirometry']]
display(rows_with_null_spirometry)
```

```python
# Filter the DataFrame to include Patient_ID, Visit_ID, and Spirometry for patients with null Spirometry values
rows_with_null_spirometry = df[df['Patient_ID'].isin(patient_ids_with_null_spirometry)][['Patient_ID', 'Visit_ID', 'Spirometry']]

# Display the updated DataFrame
display(rows_with_null_spirometry)
```

##Ventilation

```python
print(df['Ventilation'].unique())
```

```python
# Create a bar plot for Ventilation
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Ventilation')
plt.title('Distribution of Ventilation')
plt.xlabel('Ventilation (0: No, 1: Yes)')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['No', 'Yes'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
ventilation_subset = df[df['Ventilation'].isin([2.0, 3.0])]
display(ventilation_subset['Ventilation'].unique())
display(ventilation_subset['Ventilation'].value_counts())
```

##Age

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Age_at_onset
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age_at_onset', bins=20, kde=True)
plt.title('Distribution of Age at Onset')
plt.xlabel('Age at Onset')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Box plot for Age_at_onset
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='Age_at_onset')
plt.title('Box Plot of Age at Onset')
plt.ylabel('Age at Onset')
plt.grid(True)
plt.show()
```
