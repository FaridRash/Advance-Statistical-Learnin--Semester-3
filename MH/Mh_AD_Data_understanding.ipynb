#Feature info

```python
"""
Patient_ID – unique identifier for each patient.

Visit_ID – unique identifier for each clinical visit (patients can have multiple visits).

Date_of_birth – patient’s date of birth.

Symptom_onset_date – date when symptoms first appeared.

Diagnosis_date – date of confirmed diagnosis.

Sex – patient’s gender. 0 = male, 1 = female.

Age_at_onset – age when symptoms started.

Clinical_onset_type – type of symptom onset (e.g., limb, bulbar, respiratory).

Spirometry – lung function measure (often FVC%).

FVC – Forced Vital Capacity (key respiratory measure).

Ventilation – whether patient uses ventilatory support (yes/no).

Tracheostomy – whether patient has undergone tracheostomy.

PEG – whether patient has a percutaneous endoscopic gastrostomy (feeding tube).

Family_history – family history of the disease (yes/no).

Therapy – treatments patient is receiving.

Diagnostic_delay – time between symptom onset and diagnosis (months).

MRC_Upper_Limb – Medical Research Council score for upper limb strength.

MRC_Lower_Limb – MRC score for lower limb strength.

MRC_Bulbar – MRC score for bulbar function (speech/swallowing).

PUMNS_Upper_Limb – Penn Upper Motor Neuron Score for upper limb.

PUMNS_Lower_Limb – Penn UMN Score for lower limb.

PUMNS_Bulbar – Penn UMN Score for bulbar region.

CNS_Linear_Score – Composite Neurological Score (linear scale).

KINGS_Total – King’s clinical staging score (disease progression stage).

Disease_duration – time since symptom onset (months or years).

Progression_rate – rate of disease worsening (e.g., ALSFRS decline per month).

ALSFRS_R – Revised ALS Functional Rating Scale (0–48; higher = better function).

ALSAQ_5 – ALS Assessment Questionnaire, 5-item summary (quality of life).
"""
```
