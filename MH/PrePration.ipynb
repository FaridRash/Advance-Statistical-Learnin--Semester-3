#Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Reading

```python
df = pd.read_csv('/content/Advance-Statistical-Learnin--Semester-3/Dataset/df_processed.csv')
display(df.head())
```

## EDA

```python
most_frequent_patient_id = df['Patient_ID'].value_counts().idxmax()
most_frequent_patient_rows = df[df['Patient_ID'] == most_frequent_patient_id]
display(most_frequent_patient_rows)
```

```python
variables_to_plot = ['Age_at_onset', 'Diagnostic_delay', 'MRC_Upper_Limb', 'MRC_Lower_Limb', 'MRC_Bulbar', 'PUMNS_Upper_Limb', 'PUMNS_Lower_Limb', 'PUMNS_Bulbar', 'CNS_Linear_Score', 'Disease_duration', 'Progression_rate', 'ALSFRS_R', 'ALSAQ_5']
for var in variables_to_plot:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df, x=var, y='ALSFRS_R')
    plt.title(f'Scatter plot of {var} vs ALSFRS_R')
    plt.xlabel(var)
    plt.ylabel('ALSFRS_R')
    plt.show()
```

```python
variables_to_correlate = ['Age_at_onset', 'Diagnostic_delay', 'MRC_Upper_Limb', 'MRC_Lower_Limb', 'MRC_Bulbar', 'PUMNS_Upper_Limb', 'PUMNS_Lower_Limb', 'PUMNS_Bulbar', 'CNS_Linear_Score', 'Disease_duration', 'Progression_rate', 'ALSFRS_R', 'ALSAQ_5']

correlation_matrix_spearman = df[variables_to_correlate].corr(method='spearman')

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_spearman, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Correlation Matrix')
plt.show()

#we use it because out variables dose not have any liniear correlation .
```

```python
alsaq5_visit_34 = df[df['Visit_ID'] == 34]['ALSAQ_5']
display(alsaq5_visit_34)
```

```python
# Sort the DataFrame by 'ALSAQ_5' in ascending order
df_sorted_alsaq5 = df.sort_values(by='ALSAQ_5', ascending=True)

# Get the top 5 lowest ALSAQ_5 values
lowest_5_alsaq5 = df_sorted_alsaq5.head(5)

# Display the number of visits with the 5 lowest ALSAQ_5 values
num_visits_lowest_5 = len(lowest_5_alsaq5)
print(f"Number of visits with the 5 lowest ALSAQ_5 values: {num_visits_lowest_5}")

# Optionally, display the visits with the 5 lowest ALSAQ_5 values
display(lowest_5_alsaq5)
```
